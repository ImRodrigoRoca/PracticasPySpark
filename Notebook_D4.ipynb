### 3. RDDs: Transformaciones 	clave – valor (K, V)

```
# RDDs de pares clave-valor (K, V)

# Crear RDD
palabras = sc.parallelize(['HOLA', 'Que', 'TAL', 'Bien'])

# Aplicar transformación 
pal_long = palabras.map(lambda elem: (elem, len(elem)))

# Imprimir resultado
pal_long.collect()
```

##### reduceByKey(f): 

- Al llamarlo sobre un RDD de pares clave-valor (K, V), devuelve otro de pares (K,V) donde los valores de cada clave se han agregado usando la función dada.

```
# Crear RDD
r = sc.parallelize([('A', 1), ('C', 4), ('A', 1), ('B', 1), ('B', 4)])

# Aplicar transformación 
rr = r.reduceByKey(lambda v1, v2: v1 + v2)

# Imprimir resultado
print(rr.collect())
```

```
# Crear RDD
r = sc.parallelize([('A', 1), ('C', 4), ('A', 1), ('B', 1), ('B', 4)])

# Aplicar transformación 
rr = r.reduceByKey(lambda v1, v2: str(v1 + v2))

# Imprimir resultado
print(rr.collect())
```

##### groupByKey(f): 

- Al llamar lo sobre un RDD de pares clave-valor (K, V), devuelve otro de pares (K, seq[V]) donde los valores de cada clave se han convertido a una secuencia.

```
# Crear RDD
r = sc.parallelize([('A', 1), ('C', 2), ('A', 3), ('B', 4), ('B', 5)])

# Imprimir resultados
print(r.groupByKey().mapValues(list).collect())
print(r.groupByKey().mapValues(len).collect())
```

##### sortByKey(): 

- Ordena un RDD de pares clave-valor (K, V) por clave.

```
# Crear RDD
rdd = sc.parallelize([('A', 1), ('B', 2), ('C', 3), ('A', 4), ('A', 5), ('B', 6)])

# Aplicar transformación 
res = rdd.sortByKey(False)

# Imprimir resultado
print(res.collect())
```

##### join(rdd): 

- Hace un join de dos rdd de pares (K, V1) y (K, V2) y devuelve otro RDD con claves (K, (V1,V2)).

```
# Crear RDD
rdd1 = sc.parallelize([('A', 1), ('B', 2), ('C', 3)])
rdd2 = sc.parallelize([('A', 4), ('B', 5), ('C', 6)])

# Aplicar transformación
rrdjoin = rdd1.join(rdd2)

# Imprimir resultado
print(rrdjoin.collect())
```

```
# Consideraciones sobre “join()”
# Crear RDD
rdd1 = sc.parallelize([('A', 1), ('B', 2), ('C', 3)])
rdd2 = sc.parallelize([('A', 4), ('B', 5), ('C', 6), ('D', 7)])

# Aplicar transformación
rddjoin = rdd1.join(rdd2)
rddleftOuterJoin = rdd1.leftOuterJoin(rdd2)
rddrightOuterJoin = rdd1.rightOuterJoin(rdd2)
rddfullOuterJoin = rdd1.fullOuterJoin(rdd2)

# Imprimir Join(). Imprime los valores que coinciden entre ambos RDD
print(rrdjoin.collect())

# Imprimir leftOuterJoin. Imprime todos los valores del rdd1 y los que coinciden con el rdd2
print(rrdleftOuterJoin.collect())

# Imprimir rightOuterJoin. Imprime todos los valores del rdd2 y los que coinciden con el rdd1
print(rddrightOuterJoin.collect())

# Imprimir fullOuterJoin. Imprime todos los valores de los rdd1 y rdd2, no importa si coinciden o no.
print(rddfullOuterJoin.collect())
```
